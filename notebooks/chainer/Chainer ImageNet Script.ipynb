## Download ImageNet dataset

```bash
curl -LO http://www.vision.caltech.edu/Image_Datasets/Caltech101/101_ObjectCategories.tar.gz
tar zxvf 101_ObjectCategories.tar.gz
```

## Pre-process the dataset

Create the train.txt and test.txt

```bash
python3 make_train_data.py 101_ObjectCategories
```

## Crop all the images

```bash
python3 crop.py images/ images/
```

## Compute mean image

```bash
python3 ../chainer/examples/imagenet/compute_mean.py train.txt
```

## Train a model


```bash
python3 ../chainer/examples/imagenet/train_imagenet.py --gpu -1 --epoch 1  --mean mean.npy train.txt test.txt
```
